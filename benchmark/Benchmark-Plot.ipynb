```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df1 = pd.read_csv('docker_all.csv_output.csv', names=list(range(1,500)))
df1=df1.set_index(1)

# Create a figure instance
fig = plt.figure(1, figsize=(9, 6))

# Create an axes instance
ax = fig.add_subplot(111)
ax.boxplot(df1.T.dropna().values)
ax.set_xticklabels(list(df1.index.values),rotation=90)
ax.set_ylabel('Time Taken')
```
